```python
import numpy as np
np.random.seed(123)  

for i in range(1, 4):
    print(f"MATRIX {i}")
    print('='*50)
    
    # random 3x3 matrix with integers from -10 to 10
    M = np.random.randint(-10, 11, size=(3, 3))
    print(f"\nM{i} =")
    print(M)
    
    # inverse using numpy package
    det = np.linalg.det(M)
    print(f"\nDeterminant = {det:.4f}")
    
    if abs(det) < 1e-10:
        print("Matrix is singular - no inverse exists!")
    else:
        M_inv = np.linalg.inv(M)
        print(f"\nM{i}^(-1) =")
        print(np.round(M_inv, 4))
        
        # Verification
        product = M @ M_inv
        print(f"\nM{i} × M{i}^(-1) =")
        print(np.round(product))
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("kc_house_data.csv")

# Drop columns to ignore
df = df.drop(columns=["id", "date", "zipcode"])

# Compute statistics
stats = pd.DataFrame({
    "average": df.mean(),
    "min": df.min(),
    "max": df.max(),
    "variance": df.var()
})

# Identify features with lowest/highest average
lowest_avg_feature = stats["average"].idxmin()
highest_avg_feature = stats["average"].idxmax()

# Identify features with lowest/highest variance
lowest_var_feature = stats["variance"].idxmin()
highest_var_feature = stats["variance"].idxmax()

# Display results
print("Summary statistics:\n")
display(stats)

print("\nFeature with lowest average:")
print(f"{lowest_avg_feature}: {stats.loc[lowest_avg_feature, 'average']}")

print("\nFeature with highest average:")
print(f"{highest_avg_feature}: {stats.loc[highest_avg_feature, 'average']}")

print("\nFeature with lowest variance:")
print(f"{lowest_var_feature}: {stats.loc[lowest_var_feature, 'variance']}")

print("\nFeature with highest variance:")
print(f"{highest_var_feature}: {stats.loc[highest_var_feature, 'variance']}")
```
